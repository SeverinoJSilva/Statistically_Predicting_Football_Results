```python
cols = ['Home_team_score','Away_team_score']
pl[cols] = pl[cols].astype(int)
```

```python
Home = pl.groupby('Home_team_name').mean(numeric_only=True)
Home.columns=['Make_Home', 'Take_Home']
Home
```

```python
Away = pl.groupby('Away_team_name').mean(numeric_only=True)
Away.columns=['Take_Away', 'Make_Away']
Away
```

```python
tb_data = Home.merge(Away, left_index=True, right_index=True)
tb_data = tb_data.reset_index()
tb_data.rename(columns={'Home_team_name':'Team'}, inplace=True)
tb_data
```

```python
from scipy.stats import poisson

# Select the Team
home_team = 'Manchester United'
away_team = 'Manchester City'

lambda_home = tb_data.loc[tb_data['Team']== home_team,'Make_Home'].iloc[0] * tb_data.loc[tb_data['Team']== away_team,'Take_Away'].iloc[0]
lambda_away = tb_data.loc[tb_data['Team']== away_team,'Make_Away'].iloc[0] * tb_data.loc[tb_data['Team']== home_team,'Take_Home'].iloc[0]
    
pv_home = 0
pv_draw = 0
pv_away = 0

# show the home or visitor probability of winning or draw
for i in range(9):
    for j in range(9):
        p_result = poisson.pmf(i, lambda_home) * poisson.pmf(j,lambda_away)
        if i == j:
            pv_draw += p_result
        elif i > j:
            pv_home += p_result
        elif i < j:
            pv_away += p_result
print('Probability:')            
print(f'{home_team} =>| {round(pv_home,2)*100} % | {round(pv_draw,3)*100} % | {round(pv_away,2)*100} % |<= {away_team}\n')

# show the goals probability
print('Probability of goals in the match')        
for t_goals in range(9):
    p_t_goals = 0
    for i in range(t_goals + 1):
        j = t_goals - i
        if j < 9: 
            p_t_goals += poisson.pmf(i, lambda_home) * poisson.pmf(j, lambda_away)
    
    print(f'number: {t_goals} gols: {p_t_goals * 100:.2f}%')
    
# visulization result
print('\nIf the game happened, see the result:')  
display(pl.query(f"Home_team_name=='{home_team}' & Away_team_name=='{away_team}'"))
```
